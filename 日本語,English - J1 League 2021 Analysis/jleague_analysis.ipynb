# DATA VISUALIZATION
# データビジュアライゼーション

```python
df['シュート数/ゴール数'] = df['シュート'] / df['得点']

central_tendency = {}
central_tendency['Mean'] = df['シュート数/ゴール数'].mean().round()
central_tendency['Median'] = df['シュート数/ゴール数'].median().round()
central_tendency['Mode'] = df['シュート数/ゴール数'].mode()[0].round()

central_tendency
```

We can see that the average Shots / Goal Scored is 8. Let's see each team individually

シュート数/ゴール数の平均は8、各チームを個別に見てみましょう。

```python

plt.figure(figsize=(10,10))
sns.barplot(x = df['チーム名'],
            y = df['シュート数/ゴール数'])
plt.title('シュート数 / ゴール数')
plt.xticks(
    rotation=45,
    horizontalalignment='right',
    fontweight='light',
    fontsize='medium',)
plt.show()

```

Unsurprisingly, the top 3 teams were the ones with the lowest shot/goal ratio. Surprisingly though, the team with the highest shot/goal ratio was Kashiwa Reysol who actually managed to avoid relegation.

当然のことですが、上位3チームはシュート数／ゴール数が最も少ないチームであることが分かります。また意外なことにも、降格を免れた柏レイソルはシュート数／ゴール数が最も多かったチームでした。

Now let's look at the how many minutes it takes for each team to get a shot off, first league-wide

チームがシュートを打つまでに何分かかっているのかを見てみます

```python
df['試合時間/シュート数'] = df['試合時間'] / df['シュート']

central_tendency = {}
central_tendency['Mean'] = df['試合時間/シュート数'].mean().round()
central_tendency['Median'] = df['試合時間/シュート数'].median().round()
central_tendency['Mode'] = df['試合時間/シュート数'].mode()[0].round()

central_tendency
```

Looks like the league average is 10 minutes for each shot. Now let's look at each team individually

シュートを打つまで平均10分かかっています。各チームを個別に見ていきます。

```python
df['試合時間/シュート数'] = df['試合時間'] / df['シュート']

plt.figure(figsize=(10,10))
sns.scatterplot(x = df['チーム名'],
               y = df['試合時間/シュート数'],
               size=df['試合時間/シュート数'],
               sizes=(100,600),
               hue=df['試合時間/シュート数'])
plt.title('シュート数 / ゴール数')
plt.xticks(
    rotation=45,
    horizontalalignment='right',
    fontweight='light',
    fontsize='medium',)
plt.show()


```

Looks like the second place team Yokohama F.Marinos is actually the team that shoot most frequently. Surprising to note that the 3rd place team Vissel Kobe is acutually the 8th team with the most shots showing how effecient they were. Meanwhile, the Consadole Sapporo, the 4th team with the most shots, actually finished in 10th place showing that they perhaps lacked some efficiency in front of the net.

2位の横浜F・マリノスは、シュート数が多いチームです。3位のヴィッセル神戸は、シュート数で8位と、効率的なチームであったことがります。一方、シュート数4位のコンサドーレ札幌は10位と、ネット前でのは非効率だった。

Let's look at the % of shots against that found the back of the net

シュート数のうち、ゴールを決めた割合を見てみます。

```python
df['失点/被シュート'] = df['失点'] / df['被シュート'] * 100

central_tendency = {}
central_tendency['Mean'] = df['失点/被シュート'].mean().round()
central_tendency['Median'] = df['失点/被シュート'].median().round()
central_tendency['Mode'] = df['失点/被シュート'].mode()[0].round()

central_tendency
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x = df['チーム名'],
               y = df['失点/被シュート'],
               size=df['被シュート'],
               sizes=(100,600),
               hue=df['被シュート'])
plt.title('シュート数 / ゴール数')
plt.xticks(
    rotation=45,
    horizontalalignment='right',
    fontweight='light',
    fontsize='medium',)
plt.show()
```

What's surprising here is that Kashima Antlers, the 4th placed team, had above average shots against/goals percentage,　while everybody in the top 8 had below average. This is balanced out by the fact that they were also the team that suffered the least amount of shots which helped them finish in such a high position.

意外なことに、4位の鹿島アントラーズは失点/被シュート数で平均を上回ってました。これは、鹿島アントラーズが最も被シュート数が少なかったこととバランスが取れており、最終的には上位に食い込んでいます。

Now let's look at fouls and cards

次に反則とカードについて見て見てみます。

```python
df['試合時間/反則'] = df['試合時間'] / df['反則']

central_tendency = {}
central_tendency['Mean'] = df['試合時間/反則'].mean().round()
central_tendency['Median'] = df['試合時間/反則'].median().round()
central_tendency['Mode'] = df['試合時間/反則'].mode()[0].round()

central_tendency
```

```python
plt.figure(figsize=(10,10))
sns.barplot(x = df['チーム名'],
            y = df['試合時間/反則'])
plt.title('試合時間/反則')
plt.xticks(
    rotation=45,
    horizontalalignment='right',
    fontweight='light',
    fontsize='medium',)
plt.show()
```

Looks like it most teams are close to the average with the outliers being Sagan Tosu which commit a foul about every 6 minutes and at the other end Shimizu S-Pulse who takes around 9 minutes to commit a foul   

ほとんどのチームが平均値に近いです。異常値としては、サガン鳥栖が約6分に1回、清水エスパルスが約9分に1回ファウルを犯しているようです。  

Let's look at how many fouls it takes for a team to get a card

チームがカードをもらうまでに反則数をみてみます。

```python
# Add the Yellow and Red Card categories together
# 警告と退場を合算する

sum_column = df["警告"] + df["退場"]
df["カード"] = sum_column
```

```python
df['反則/カード'] = df['反則'] / df['カード']

central_tendency = {}
central_tendency['Mean'] = df['反則/カード'].mean().round()
central_tendency['Median'] = df['反則/カード'].median().round()
central_tendency['Mode'] = df['反則/カード'].mode()[0].round()

central_tendency
```

```python
plt.figure(figsize=(10,10))
sns.barplot(x = df['チーム名'],
            y = df['反則/カード'])
plt.title('反則/カード')
plt.xticks(
    rotation=45,
    horizontalalignment='right',
    fontweight='light',
    fontsize='medium',)
plt.show()
```

Looks like Kawasaki Frontale is the team that fouls the most before receiving a card. Meanwhile, Avispa Fukuoka is the team with the worst discipline in the league recieving the most cards

川崎フロンターレは、カードをもらう前に最も多くファウルをしたチームであることが分かります。一方、アビスパ福岡は、リーグで最も多くのカードを受け取っている改善すべきチームです。
